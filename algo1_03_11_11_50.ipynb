```python
def pont_helyzete(center, pont, sugar):
  # r^2 = (x-u)^2 + (y-v)^2 (C(u;v) középpontú r sugarú kör egyenlete.)
  tavolsag = (pont[0]-center[0])**2 + (pont[1]-center[1])**2
  if tavolsag < sugar**2:
    print("A pont a körvonalon belül helyezkedik el.")
  elif tavolsag > sugar**2:
    print("A pont a körvonalon kívül helyezkedik el.")
  else: # tavolsag == sugar**2:
    print("A pont a körvonalon helyezkedik el.")
```

```python
center = [0, 0] # u,v
pont = [1, 2] # x,y
sugar = 5
pont_helyzete(center, pont, sugar)
pont_helyzete([1,1], [2,5], 3)
```

```python
def haromszog_fajtai(a, b, c):
  oldalak = sorted([a, b ,c]) # növekvő sorrendbe rendezem az értékeket
  a, b ,c = oldalak # a lesz a legkisebb, c legnagyobb
  if a + b > c:
    if a == c: # a == b == c
      print("Ez egy egyenlő oldalú háromszög")
    elif a == b or b == c:
      print("Ez egy egyenlő szárú háromszög")
    else:
      print("Ez egy általános háromszög")

    if a**2 + b**2 > c**2:
      print("Ez a háromszög hegyesszögyű")
    elif a**2 + b**2 < c**2:
      print("Ez a háromszög tompaszögű")
    else: # a**2 + b**2 == c**2
      print("Ez a háromszög derékszögű")
  else:
    print("A háromszög nem szerkeszthető")

# Bonyolultság: 2^0 + 3 x 2^1 + 3 x 2^1 + 2^0 = 1 + 6 + 6 + 1 = 14
```

```python
# D > 0 --> Két különböző valós gyök van.
# D = 0 --> Egy valós (kettős) gyök van.
# D < 0 --> Nincs valós gyök.
```

```python
import math
def masodfoku_egyenlet_megoldo(a, b, c):
  d = b**2 - 4 * a * c # Diszkrimináns kiszámítása
  if d > 0:
    x1 = (-b + math.sqrt(d) / (2 * a))
    x2 = (-b - math.sqrt(d) / (2 * a))
    print(f"d > 0: Két különböző valós gyök van, x1={x1} és x2={x2}")
  elif d == 0:
    x = -b / (2 * a)
    print(f"d = 0: Egy valós (kettős) gyök van, x={x}")
  else:
    print("d < 0: Nincs valós gyök.")
```

```python
masodfoku_egyenlet_megoldo(2, 3, -4)
```
